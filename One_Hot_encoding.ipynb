```python
data = ['cold', 'cold', 'warm', 'cold', 'hot', 'hot', 'warm', 'cold', 'warm', 'hot']
print(data)

#encoding categories
data=le.fit_transform(data)
print(data)
```

```python
ohe=OneHotEncoder(sparse=False)
data=data.reshape(len(data),1)
ohe=ohe.fit_transform(data)
print(ohe)
```

```python
#data = ['cold', 'cold', 'warm', 'cold', 'hot', 'hot', 'warm', 'cold', 'warm', 'hot']
#invert first example
inverted=le.inverse_transform([argmax(ohe[0:])])
print(inverted)
```

```python
inverted=le.inverse_transform([argmax(ohe[5:])])
print(inverted)
```
